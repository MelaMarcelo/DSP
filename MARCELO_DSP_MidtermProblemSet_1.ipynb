Part 1. [200 pts] Compute and plot (you may or may not use software) the following complex numbers. You should have a total of three (3) graphs. First graph includes numbers 1 to 4. Second graph includes numbers 5 to 9. The third graph includes only number 10.

1. $ z_1 = 1-5i $
2. $ z_2 = 3+2i $
3. $ z_3 = -7+i $
4. $ z_4 = -1.6i $
5. $ z_5 = z_1 \cdot z_{1}^{*} $
6. $ z_6 = z_2 \cdot (z_1-z_3)^* $
7. $ z_7 = \frac{z_4}{z_1} $
8. $ z_8 = \frac{z_4}{z_2} (z_1+z_4) $
9. $ z_9 = \frac{z_1}{z_3} + \frac{2z_4}{3} $
10. $ \omega = e^{z_4} $

```python
#Z1-Z4

Z_1 = complex(1, -5)
Z_2 = complex(3, 2)
Z_3 = complex(-7, 1)
Z_4 = complex(0, -1.6)

Real = [zx.real for zx in [Z_1, Z_2, Z_3, Z_4]]
Imaginary = [zx.imag for zx in [Z_1, Z_2, Z_3, Z_4]]

fig, ax = plt.subplots(figsize=(7, 7))

ax.scatter(Real, Imaginary, color='Red')
ax.set_title("Plot for Z1, Z2, Z3, Z4")
ax.set_xlabel('Real')
ax.set_ylabel('Imaginary')
ax.set_xlim((-10, 10))
ax.set_ylim((-10, 10))

for i in range(4):
    ax.vlines(Real[i], 0, Imaginary[i], 'blue', ':')
    ax.hlines(Imaginary[i], Real[i], 0, 'green', ':')
    ax.text(Real[i], Imaginary[i], f'Z{i + 1} = {Real[i]} + {Imaginary[i]}i')

ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid()
plt.show()
```

```python
#Z5-Z9

Z_1_conj = np.conj(Z_1)
Z_1Z_3 = Z_1 - Z_3
Z_1Z_3_conj = np.conj(Z_1Z_3)

Z_5 = Z_1 * Z_1_conj
Z_6 = Z_2 * Z_1Z_3_conj
Z_7 = Z_4 / Z_1
Z_8 = (Z_4 / Z_2) * (Z_1 + Z_4)
Z_9 = (Z_1 + Z_3) + ((2 * Z_4) / 3)

Real2 = [zx.real for zx in [Z_5, Z_6, Z_7, Z_8, Z_9]]
Imaginary2 = [zx.imag for zx in [Z_5, Z_6, Z_7, Z_8, Z_9]]

fig, ax = plt.subplots(figsize=(7, 7))

ax.scatter(Real2, Imaginary2, color='Red')
ax.set_title("Plot for Z5, Z6, Z7, Z9, Z9")
ax.set_xlabel('Real')
ax.set_ylabel('Imaginary')
ax.set_xlim((-30, 40))
ax.set_ylim((-30, 40))

for i in range(5):
    ax.vlines(Real2[i], 0, Imaginary2[i], 'blue', ':')
    ax.hlines(Imaginary2[i], Real2[i], 0, 'green', ':')
    ax.text(Real2[i], Imaginary2[i], f'Z{i + 5}')

    #with value after Z5..  it covered the lines so nah
    #ax.text(Real2[i], Imaginary2[i], f'Z{i + 5} = {Real2[i]} + {Imaginary2[i]}i')

ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid()
plt.show()
```

Part 2. [50 pts] Compute for the Magnitude and Phase of Part 1 numbers 1-9. You should denote the magnitude as $M_x$ and Phase as $\phi _x$ where 𝑥 is the variable name of the complex number.

```python
def Magnitude_Phase(Z):
    Magnitude = abs(Z)
    Angle = np.angle(Z)
    Phase = np.rad2deg(Angle)
    return Magnitude, Phase
```

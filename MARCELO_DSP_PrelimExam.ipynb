# **MAGNITUDE AND ANGLE OF THE LINEAR TRANSFORMATIONS**

```python
#from lecture notes **Vector Norm**
def calc_linalg_norm(n, vector):
    n_sum = np.sum(vector**n)
    vect_norm = np.power(n_sum,1/n)
    return vect_norm

def l1_norm(vector):
    vect_norm = np.sum(vector)
    return vect_norm
```

```python
h = H
magnitude = np.linalg.norm(h)
print(h)
```

```python
y = Y
magnitude = np.linalg.norm(y)
print(y)
```

```python
o = O
magnitude = np.linalg.norm(o)
print(o)
```

```python
#1 H ⋅ Y (Magnitude)
num_1 = H@Y
magnitude = np.linalg.norm(num_1)
print(num_1)
print(f"Magnitude: {calc_linalg_norm(2, num_1)}")
```

```python
#1 H ⋅ Y (Angle)
num1_angle = H @ Y

angle_radians = np.arctan2(num1_angle[1, 0], num1_angle[0, 0])
angle_degrees = np.degrees(angle_radians)

print("Angle of the linear transformation (in degrees):", angle_degrees)
```

**Part 3**. [50 pts] Determine whether the resulting linear transformations are linearly dependent. Provide
necessary solutions both numerically using determinants (hand-solved) and computationally (using a
Python program).

1.
\begin{equation}
\begin{pmatrix} 5 & 0 & 0 \\ 0 & 5 & 0 \\ 0 & 0 & 5 \end{pmatrix} ⋅
%
\begin{pmatrix} 1 & 0 & 5 \\ 2 & 7 & 6 \\ 6 & 4 & 7 \end{pmatrix}
\end{equation}
<br>
2.
\begin{equation}
\begin{pmatrix} 1 & 2 & 6 \\ 3 & 15 & 4 \\ 2 & 10 & 3 \end{pmatrix} ⋅
%
\begin{pmatrix} 5 & 2 & 4 \\ 6 & 2 & 4 \\ 0 & 1 & 1 \end{pmatrix}
\end{equation}

```python
#Put the matrices separately to code it easier
#1
A = np.array([
    [5,0,0],
    [0,5,0],
    [0,0,5]
])

B = np.array([
    [1,0,5],
    [2,7,6],
    [6,4,7]
])

#2
C = np.array([
    [1,2,6],
    [3,15,4],
    [2,10,3]
])

D = np.array([
    [5,2,4],
    [6,2,4],
    [0,1,1]
])
```

```python
num1_AB = np.dot(A, B)
det_num1_AB = np.linalg.det(num1_AB)

if det_num1_AB != 0:
    print("Linearly Dependent")
    print("Determinant:", det_num1_AB)
else:
    print("Not linearly Dependent")
```

```python
num2_CD = np.dot(C, D)
det_num2_CD = np.linalg.det(num2_CD)

if det_num2_CD != 0:
    print("Linearly Dependent")
    print("Determinant:", det_num2_CD)
else:
    print("Not linearly Dependent")
```

is the vector corresponding to the amplitudes of the signal.

\begin{equation}
\begin{bmatrix} T = 0, \pi{/4}, \pi{/2}, 3\pi{/4}, \pi \end{bmatrix} ;

\begin{bmatrix} G = (5, 3, 0, -3, 5) \end{bmatrix}

```python
T = np.array([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi])
G = np.array([5, 3, 0, -3, 5])

#Calculate
PSignal = G * np.sin(T)

#Plotting
plt.plot(T, PSignal, marker='o')
plt.xlabel('Time (T)')
plt.ylabel('Amplitude (G)')
plt.title('Signal Corresponding to Given T&G Vectors')
plt.grid(True)
plt.show()
```
